```python
class Fraction:
  def __init__(self,n,d):
    self.num=n
    self.den=d
  def __str__(self): # how an object is converted into human-readable string
    return "{}/{}".format(self.num,self.den)
  def __add__(self,other):
    return f"{((other.den*self.num)+(self.den*other.num))}/{self.den*other.den}"
  def __sub__(self,other):
    return f"{((other.den*self.num)-(self.den*other.num))}/{self.den*other.den}"
  def __mul__(self,other):
    return f"{(other.num*self.num)}/{self.den*other.den}"
  def __truediv__(self,other):
    return f"{self.num*other.den}/{self.den*other.num}"
```

```python
x=Fraction(4,5)
```

```python
print(x)
```

```python
y=Fraction(2,5)
```

```python
print(x+y)
```

```python
print(x-y)
```

```python
print(x*y)
```

```python
print(x/y)
```
