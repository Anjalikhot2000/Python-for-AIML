**Indentation**

In Python, indentation is mandatory and is used to define blocks of code, where many other languages use braces ({}). It is a core part of the syntax, not just a matter of style.

```python
name="xyz"
if name=="xyz":
  print("line1")
  print("line2")

else:
  print("line3")
```
