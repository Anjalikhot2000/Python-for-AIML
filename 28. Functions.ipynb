# Argements - The values that are declared within a function when the function is called are known as an argument.
# Paramenters- The variables that are defined when the function is declared are known as parameters.

Types of Arguments
1) Default: Default argument is fallback value
In Python, a default parameter is defined with a fallback value as a default argument. Such parameters are optional during a function call. If no argument is provided, the default value is used, and if an argument is provided, it will overwrite the default value.

2) Positional: When you call a function with arguments without using keywords, they are called positional arguments.
Positional arguments must be in the correct order.

3) Keyword: You can send arguments with the key = value syntax.

4) Arbitrary: If you do not know how many arguments that will be passed into your function, add a * before the parameter name in the function definition.

```python
# default
def power(a=3,b=1):
  return a**b
```

```python
power(2,3)
```

```python
# Arbitrary
def flexi(*numb):
  print(type(numb))
  product=1
  for i in numb:
    print("i:",i)
    product*=i
    print("prod:",product)
  return product
```

```python
res=flexi(1,2,7)
res
```

NOTE:

We can use global variable locally but cannot change value of global variable locally.

Local: Local variables are defined inside a function and exist only during its execution. They cannot be accessed from outside the function.

Global: Global variables are declared outside all functions and can be accessed anywhere in the program, including inside functions.

```python
def f(x):
  x=x+1
  print("in f(x):x=",x)
  return x

x=3
z=f(x)
print("in main scope:z=",z)
print("in main scope:x=",x)
```

```python
# Nested function: Function within function
def f():
  print("Inside f")
  def g():
    print("Inside g")
    f()
  g()
f()
```

```python
def f():
  print("Inside f")
  def g():
    print("Inside g")
  g()
f()
```

```python
def g(x):
  def h():
    x="abc"
  x=x+1
  print("in g(x):x=",x)
  h()
  return x

x=3
z=g(x)
print(z)
```

```python
def g(x):
  def h(x):
    x=x+1
    print("in h(x):x=",x)
  x=x+1
  print("in g(x):x=",x)
  h(x)
  return x
x=3
z=g(x)
print("in main scope:x=",x)
print("in main scope:z=",z)
```
