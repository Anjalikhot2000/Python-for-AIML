```python
# Lambda function
# - Anonymous function
# Syntax
# lambda input:expression
```

```python
x=lambda x:x**2
x(9)
```

```python
a=lambda x,y:x+y
a(2,3)
```

```python
b=lambda x:x[0]=="a"
b("apple")
```

```python
# map
# map(function,iteratables)

l=[1,2,3,4,5]
a=list(map(lambda x:x*2,l))
a
```

```python
a=list(map(lambda x:"Even" if x%2==0 else "Odd",l))
```

```python
# Filter

l=[1,2,3,4,5,67,8]
a=list(filter(lambda x:x>4,l))
a
```

```python
fruits=["apple","orange","mango","guava"]
a=list(filter(lambda fruit:"e" in fruit,fruits))
a
```

```python
# reduce
import functools
l=[1,2,3,4,5,67,8]
a=functools.reduce(lambda x,y:x+y,l)
a
```

```python
functools.reduce(lambda x,y: x if x>y else y,l)
```
