```python
# Iterative Vs Recurive
# a*b
# iterative
a=5
b=4
prod=0
for i in range(b):
  prod+=a
print(prod)
```

```python
# Recursion
def mul(a,b):
  if b==1:
    return a
  else:
    return (a+mul(a,b-1))

print(mul(5,4))
```

```python
# factorial
def factorial(num):
  if num==1:
    return num
  else:
    return(num*factorial(num-1))

print(factorial(5))
```

```python
# Palindrome
def Palindrome(s):
  if len(s)<=1:
    print("palindrome")
  else:
    if s[0]==s[-1]:
      Palindrome(s[1:-1])
    else:
      print("not palindrome")

Palindrome("madam")
Palindrome("python")
Palindrome("abba")
```

```python
# The rabbit problem
# Fibonacci
def Fibonacci(a):
  if a==0 or a==1:
    return 1
  else:
    return Fibonacci(a-1)+Fibonacci(a-2)

print(Fibonacci(12))
```

```python
import time
def memo(m,d):
  if m in d:
    return d[m]
  else:
    d[m]=memo(m-1,d)+memo(m-2,d)
    return d[m]
start=time.time()
d={0:1,1:1}
print(memo(48,d))
print(time.time()-start)
print(d)
```
