Polymorphism

 Polymorphism in Python allows different classes, functions, or operators to share the same name, behaving differently based on the input or object type—meaning "many forms".

```python
# Ploymorphism

class Phone:
  def __init__(self,price,brand,camera):
    self.price=price
    self.brand=brand
    self.camera=camera
  def buy(self):
    print("Phone")

class SmartPhone(Phone):
  def buy(self):
    print("SmartPhone")
```

```python
s=SmartPhone(23000,"one plus",15)
```

```python
s.buy() # Method overriding
```
