**Keywords**
- Python is a case sensitive programing language
- A keyword is word that is reserved by a program because the word has a special meaning
- Keywords can be commands or parameeters.
- Every programming language has a set of keywords that cannot be used as variable names

```python
import keyword
print(keyword.kwlist)
```

**Identifiers**
- A python identifier is a name used to identify a variable, function,class, module or other object

- Rules for setting identifiers:
  - Can only start with alphabets or _
  - Followed by 0 or moser letter,_ and digits
  - Keywords cannot be used as an identifiers
