```python
class Atm:
  # __init__ is constructor- Constructor is a special method which is initialized automatically when object is created
  def __init__(self): # variable is declared inside __init__
    self.__pin=""
    self.__balance=0
    self.menu()

  def menu(self):
    user_input=0
    while(user_input!='5'):
      user_input=input("""Hello, how would you like to proceed?
      1. Enter 1 to create pin
      2. Enter 2 to deposit
      3. Enter 3 to withdraw
      4. Enter 4 to check balance
      5. Enter 5 to exit\n""")
      if user_input=='1':
        self.create_pin()
      elif user_input=='2':
        self.deposit()
      elif user_input=='3':
        self.withdraw()
      elif user_input=='4':
        self.check_balance()
      elif user_input=='5':
        self.exit()
      else:
        print("Invalid input")

  def create_pin(self):
    self.__pin=input("Enter your pin")
    print("Pin set successfully!!!")

  def pin_check(self):
    temp=input("Enter your pin")
    return temp

  def deposit(self):
    temp=self.pin_check()
    if temp==self.__pin:
      amt=eval(input("Enter the amount you want to deposit"))
      self.__balance+=amt
      print("Deposit successfully!!!")
    else:
      print("Incorrect pin")

  def withdraw(self):
    temp=self.pin_check()
    if temp==self.__pin:
      amt=eval(input("Enter the amount you want to withdraw"))
      if amt<self.__balance:
        self.__balance-=amt
        print(f"Rs.{amt} withdrawn successfully and remaining balance is Rs.{self.__balance}")
      else:
        print("Insufficient amount")
    else:
      print("Incorrect pin")

  def check_balance(self):
    temp=self.pin_check()
    if temp==self.__pin:
      print(f"Your balance is Rs.{self.__balance}")
    else:
      print("Incorrect pin")

  def exit(self):
    print("Thanks, visit again")

  def get_pin(self):
    return self.__pin

  def set_pin(self,new_pin):
    self.__pin=new_pin
    return self.__pin
```
