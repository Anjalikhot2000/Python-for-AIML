# Aggregation

image.png

```python
class Customer:
  def __init__(self,name,gender,address):
    self.name=name
    self.gender=gender
    self.address=address

  def edit_profile(self,new_name,new_city,new_pin,new_state):
    self.name=new_name
    self.address.change_address(new_city,new_pin,new_state)

class Address:
  def __init__(self,city,pincode,state):
    self.city=city
    self.pincode=pincode
    self.state=state

  def change_address(self,new_city,new_pin,new_state):
    self.city=new_city
    self.pincode=new_pin
    self.state=new_state

add=Address("Bangalore",123456,"Karnataka")
cust=Customer("XYZ","Female",add)

print(cust.address.city)
print(cust.address.pincode)
cust.edit_profile("Anjali","Pune",345678,"Maharashtra")
print(cust.address.state)
```

# Inheritance

- Inheritance in Python allows a child class to derive attributes and methods from a parent class, promoting code reusability and hierarchical organization.

image.png

```python
class User:
  def login(self):
    print("Login")
  def reg(self):
    print("Register")

class Student(User):# Inherting
  def enroll(self):
    print("Enroll")
  def review(self):
    print("Review")
```

```python
s1=Student()
s1.login()
s1.reg()
s1.enroll()
s1.review()
```

```python
# super() - The super() function in Python is a built-in function used within a subclass (child class) to call methods and access properties from its superclass (parent class)

class Phone:
  def __init__(self,price,brand,camera):
    self.price=price
    self.brand=brand
    self.camera=camera
  def buy(self):
    print("Phone")

class SmartPhone(Phone):
  def buy(self):
    print("SmartPhone")
    super().buy()
```

```python
s=SmartPhone(20000,"heya",17)
s.buy()
```

```python
class Phone:
  def __init__(self,price,brand,camera):
    self.price=price
    self.brand=brand
    self.camera=camera

class SmartPhone(Phone):
  def __init__(self,price,brand,camera,os,ram):
    super().__init__(price,brand,camera)
    self.os=os
    self.ram=ram

s=SmartPhone(20000,"Samsung",12,"Android",2)
print(s.os)
print(s.ram)
```

# Types of Inheritance
1) Single-level
2) Multi-level
3) Hirerachical
4) multiple
5) Hybrid - combination of above Inheritance


image.png

```python
# Single - Inheritance
class Phone:
  def __init__(self,price,brand,camera):
    self.price=price
    self.brand=brand
    self.camera=camera

class SmartPhone(Phone):
  def __init__(self,price,brand,camera,os,ram):
    super().__init__(price,brand,camera)
    self.os=os
    self.ram=ram

s=SmartPhone(20000,"Samsung",12,"Android",2)
print(s.os)
print(s.ram)
```

```python
# Multi-level

class Product:
  def review(self):
    print("Product Customer review")

class Phone(Product):
  def __init__(self,price,brand,camera):
    self.price=price
    self.brand=brand
    self.camera=camera

  def buy(self):
    print("Phone")

  def return_phone(self):
    print("Returning Phone")

class SmartPhone(Phone):
  pass

s=SmartPhone(2000,"Apple",12)
s.buy()
s.review()
p=Phone(23344,"qwerty",34)
p.review()
```

```python
# Hierarichal

class User:
  def login(self):
    print("Login")
  def reg(self):
    print("Register")

class Student(User):# Inherting
  def enroll(self):
    print("Enroll")
  def review(self):
    print("Review")

class Instructor(User):
  def instruct(self):
    print("Insstructing")

s=Student()
s.login()
s.reg()
s.enroll()
s.review()

i=Instructor()
i.login()
i.reg()
i.instruct()
```

```python
# Multiple

class Phone:
  def __init__(self,price,brand,camera):
    self.price=price
    self.brand=brand
    self.camera=camera

  def buy(self):
    print("Phone")

class Product:
  def review(self):
    print("Reviewing")

class SmartPhone(Phone,Product): # Method resolving order
  pass

s=SmartPhone(2000,"Apple",12)
s.buy()
s.review()
```
