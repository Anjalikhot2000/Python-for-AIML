```python
# Generator
# A simple way of creating iterators
```

```python
# Example
def gen_demo():
  yield "first statement"
  yield "second statement"
  yield "third statement"
```

```python
gen=gen_demo()
print(gen)
```

```python
next(gen)
```

```python
gen1=gen_demo()
```

```python
for i in gen1:
  print(i)
```

```python
def square(num):
  for i in range(1,num+1):
    yield i**2

gen=square(10)
```

```python
next(gen)
```

```python
for i in gen:
  print(i)
```

```python
# Range using generator

def mera_range(start,end):
  for i in range(start,end):
    yield i
```

```python
gen=mera_range(1,5)
```

```python
next(gen)
```

```python
for i in gen:
  print(i)
```

```python
# generator expression
gen1= (i**2 for i in range(1,100))
next(gen1)
```

```python
for i in gen1:
  print(i)
```
