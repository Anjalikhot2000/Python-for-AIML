```python
class Atm:
  # __init__ is constructor- Constructor is a special method which is initialized automatically when object is created
  def __init__(self): # variable is declared inside __init__
    self.pin=""
    self.balance=0
    self.menu()

  def menu(self):
    user_input=0
    while(user_input!='5'):
      user_input=input("""Hello, how would you like to proceed?
      1. Enter 1 to create pin
      2. Enter 2 to deposit
      3. Enter 3 to withdraw
      4. Enter 4 to check balance
      5. Enter 5 to exit\n""")
      if user_input=='1':
        self.create_pin()
      elif user_input=='2':
        self.deposit()
      elif user_input=='3':
        self.withdraw()
      elif user_input=='4':
        self.check_balance()
      elif user_input=='5':
        self.exit()
      else:
        print("Invalid input")

  def create_pin(self):
    self.pin=input("Enter your pin")
    print("Pin set successfully!!!")

  def pin_check(self):
    temp=input("Enter your pin")
    return temp

  def deposit(self):
    temp=self.pin_check()
    if temp==self.pin:
      amt=eval(input("Enter the amount you want to deposit"))
      self.balance+=amt
      print("Deposit successfully!!!")
    else:
      print("Incorrect pin")

  def withdraw(self):
    temp=self.pin_check()
    if temp==self.pin:
      amt=eval(input("Enter the amount you want to withdraw"))
      if amt<self.balance:
        self.balance-=amt
        print(f"Rs.{amt} withdrawn successfully and remaining balance is Rs.{self.balance}")
      else:
        print("Insufficient amount")
    else:
      print("Incorrect pin")

  def check_balance(self):
    temp=self.pin_check()
    if temp==self.pin:
      print(f"Your balance is Rs.{self.balance}")
    else:
      print("Incorrect pin")

  def exit(self):
    print("Thanks, visit again")



```

```python
# Pass by reference
class Customer:
  def __init__(self,name,gender):
    self.name=name
    self.gender=gender

def greet(customer):
  if customer.gender=="male":
    print("Hello!",customer.name,"Sir")
  else:
    print("Hello!",customer.name,"Mam")

cust=Customer("Anjali","female")

greet(cust)
```

```python
# Static method and static class
class Customer:
  __counter=0
  def __init__(self):
    self.sno=Customer.__counter
    Customer.__counter+=1

  @staticmethod
  def get_counter():
    return Customer.__counter

  @staticmethod
  def set_counter(new):
    if type(new)==int:
      Customer.__counter=new
    else:
      print("Invalid")
c1=Customer()
c2=Customer()
c3=Customer()
```
