**Literals**

- Literal is a raw data given in a variable. In python, there are various types of literals thary are as follows:
  1) Numeric
  2) String
  3) Boolean
  4) Special


1. Numeric Literals

```python
a=0b1010 # Binary
b=100 #decimal
c=0o310 #octal
d=0x12c #hexa

print(a,b,c,d)
```

```python
#float literals
float1=10.5
float2=10e2
float3=10e-2

print(float1,float2,float3)
```

```python
# Complex
x=1+3.14j
print(x,x.imag,x.real)
```

2. String Literals

```python
string='This is Python'
strings="This is Python"
char="C"
multiline_str="""This is multiline"""
unicode_str= u"\U0001F600"
raw_str=r"raw \n string"

print(string)
print(strings)
print(char)
print(multiline_str)
print(unicode_str)
print(raw_str)
```

3. Boolean Literals



```python
a=True+4
b=False+10

print(a,b)
```

4. Special Literals

```python
a=None
print(a)
```
