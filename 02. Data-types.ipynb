Python Supports 3 categories of data types:
- Basic Types - integer, float, complex, boolean and string
-  Container Types - List, Tuples, Sets and Dictionary
- User-defined -class

```python
# Integer
print(4)
print(1e308) # 10 to the power 8
print(1e309)
```

```python
# Float
print(4.5)
```

```python
# Boolean
print(True)
print(False)
```

```python
# Complex
print(4+5j)
```

```python
# String
print('Anjali')
print("Anjali")
print("""Anjali""")
```

```python
 # List :Mutable
print([1,2,3,4,5])
```

```python
# Tuple :Immutable
print((1,2,3,4,5))
```

```python
# sets: Unique value
print({1,1,2,3,4})
```

```python
# Dict: Key value pair
print({'name':'Anjali','gender':'Female'})
```
